### Unsupervised learning algorith : Takes in unlabeled data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

### We will be making our own data

```python
data = make_blobs(n_samples=200, n_features=2, centers= 4, cluster_std=1.8, random_state=101)
```

```python
type(data)
```

```python
data[0].shape
```

```python
sns.set_style('whitegrid')
```

```python
plt.scatter(data[0][:,0], data[0][:,1],c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

### KMeans will start by randomly assigning each point to a centeroid (the number of which we specify) of a cluster and then on the next round will re-evaluate

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

### What is the accuracy for the data that the unsupervised labels predicted. this is an unsupervised learning algorithm, but in this situation becuase we used the make_blobs function we can see if our kmeans algorithm is giving an accurate prediction. Will do so by plotting graphs

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow')
```
